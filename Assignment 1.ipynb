```python
#Q1
"""A for loop is typically used when you know beforehand how many times you
want to repeat a certain block of code. It's especially useful for iterating
over sequences like lists, arrays, or ranges, where the number of iterations
is well-defined."""
for i in range(1, 6):
    print(i)
    
"""A while loop is used when you don't know in advance how many times you need
to repeat a block of code, and the loop's continuation depends on a certain
condition being true."""
valid_input = False
while not valid_input:
    user_age = input("Enter your age: ")
    if user_age.isdigit() and 0 <= int(user_age) <= 120:
        valid_input = True
    else:
        print("Invalid input. Please enter a valid age.")
```

```python
#Q2
n=10
sum=0
product=1
for i in range(1,n+1):
    sum+=i
    product*=i
    i+=1
print(sum)
print(product)
    
```

```python
#with while-loop
n = 10
sum_of_numbers = 0
product_of_numbers = 1
i = 1

while i <= n:
    sum_of_numbers += i
    product_of_numbers *= i
    i += 1
print(sum_of_numbers)
print(product_of_numbers)
    
```

```python
#Q3
n=int(input("Enter the electricity unit generated by meter="))
bill=0
if 0<n<=100:
    bill=n*4.5
elif 100<n<=200:
    bill=(100*4.5)+(n-100)*6
elif 200<n<=300:
    bill=(100*4.5)+(100*6)+(n-200)*10
else:
    bill=(100*4.5)+(100*6)+(100*10)+(n-300)*20

print(bill)
    
    
```
